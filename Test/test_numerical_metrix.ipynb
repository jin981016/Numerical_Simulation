```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import linalg


A  = np.array([[2,1,-1,2],[4,5,-3,6],[-2,5,-2,6],[4,11,-4,8]],dtype=float)
B = np.array([5.,9.,4.,2.],dtype=float)

# x = np.linalg.solve(A,B)
# print(x)

N = len(A)
x = np.zeros(N)


# Gaussian elimination

#1. Forward elimination
for j in range(N) :
    for i in range(j+1,N):
        mij = A[i,j] / A[j,j]
        for k in range(j,N):
            A[i,k] -= mij*A[j,k]
        B[i] -= B[j]*mij 


x[N-1] = B[N-1] / A[N-1][N-1]  # 마지막 항 

#2. Backward elimination
for k in range(N-2,-1,-1):
    summ = np.dot(A[k,k+1:],x[k+1:])
    x[k] = (B[k]-summ)/A[k,k]
print(x)
```
